##2. Slice a Sublist:

```python
def get_sublist(lst, start, end):
    """
    Returns a sublist from the given list between start and end indices (inclusive).

    Parameters:
    lst (list): The original list.
    start (int): Starting index.
    end (int): Ending index.

    Returns:
    list: Sublist from start to end (inclusive).
    """
    return lst[start:end+1]
try:
    user_input = input("Enter elements separated by spaces: ")
    lst = user_input.split()

    if len(lst) == 0:
        raise ValueError

    start = int(input("Enter start index: "))
    end = int(input("Enter end index: "))

    if start < 0 or end >= len(lst) or start > end:
        raise IndexError

    result = get_sublist(lst, start, end)

    print("Original List:", lst)
    print("Sublist:", result)

except ValueError:
    print("Error: Please enter valid elements and indices.")

except IndexError:
    print("Error: Invalid index range.")
```

```python
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list of lists): The nested list to flatten.

    Returns:
    list: A flattened single-dimensional list.
    """
    flat_list = []
    for sublist in lst:
        for item in sublist:
            flat_list.append(item)
    return flat_list

try:
    user_input = input("Enter nested list elements (e.g., [[1,2],[3,4],[5]]): ")
    nested_list = eval(user_input)

    if not all(isinstance(sublist, list) for sublist in nested_list):
        raise ValueError

    result = flatten(nested_list)

    print("Nested List:", nested_list)
    print("Flattened List:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list.")
```

#2. Accessing Nested List Elements:



```python
def access_nested_element(lst, indices):
    """
    Accesses a specific element from a nested list using a list of indices.

    Parameters:
    lst (list): The nested list.
    indices (list): A list of indices to traverse the nested list.

    Returns:
    element: The element at the specified nested position.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element
try:
    user_input = input("Enter nested list (e.g., [[1,2,3],[4,5,6],[7,8,9]]): ")
    nested_list = eval(user_input)

    indices_input = input("Enter indices separated by spaces (e.g., 1 2): ")
    indices = list(map(int, indices_input.split()))

    result = access_nested_element(nested_list, indices)

    print("Nested List:", nested_list)
    print("Accessed Element:", result)

except (IndexError, TypeError):
    print("Error: Invalid indices for the given nested list.")

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list and indices.")
```

##3. Sum of All Elements in a Nested List:


```python
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list of any depth.

    Parameters:
    lst (list): The nested list containing numbers and/or sublists.

    Returns:
    int/float: The sum of all numbers in the nested list.
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
try:
    user_input = input("Enter a nested list (e.g., [[1,2],[3,[4,5]],6]): ")
    import ast
    nested_list = ast.literal_eval(user_input)

    result = sum_nested(nested_list)

    print("Nested List:", nested_list)
    print("Sum of all elements:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list containing numbers only.")
```

##4. Remove Specific Element from a Nested List;


```python
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list of lists): The nested list.
    elem: The element to remove.

    Returns:
    list: Modified nested list with all occurrences of elem removed.
    """
    result = []
    for sublist in lst:
        if isinstance(sublist, list):
            new_sublist = [x for x in sublist if x != elem]
            result.append(new_sublist)
        else:
            if sublist != elem:
                result.append(sublist)
    return result
try:
    user_input = input("Enter a nested list (e.g., [[1,2],[3,2],[4,5]]): ")

    import ast
    nested_list = ast.literal_eval(user_input)

    elem_input = input("Enter the element to remove: ")
    try:
        elem = int(elem_input)
    except ValueError:
        try:
            elem = float(elem_input)
        except ValueError:
            elem = elem_input

    result = remove_element(nested_list, elem)

    print("Original List:", nested_list)
    print("Modified List:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list.")
```

##5. Find the Maximum Element in a Nested List:

```python
def find_max(lst):
    """
    Finds the maximum element in a nested list of any depth.

    Parameters:
    lst (list): The nested list containing numbers and/or sublists.

    Returns:
    int/float: The maximum element in the nested list.
    """
    current_max = float('-inf')
    for item in lst:
        if isinstance(item, list):
            sub_max = find_max(item)
            if sub_max > current_max:
                current_max = sub_max
        else:
            if item > current_max:
                current_max = item
    return current_max
try:
    user_input = input("Enter a nested list (e.g., [[1,2],[3,[4,5]],6]): ")

    import ast
    nested_list = ast.literal_eval(user_input)

    result = find_max(nested_list)

    print("Nested List:", nested_list)
    print("Maximum Element:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list containing numbers only.")
```

##6. Count Occurrences of an Element in a Nested List:

```python
def count_occurrences(lst, elem):
    """
    Counts the occurrences of a specific element in a nested list.

    Parameters:
    lst (list): The nested list containing elements and/or sublists.
    elem: The element to count.

    Returns:
    int: Number of times elem occurs in the nested list.
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
            if item == elem:
                count += 1
    return count
try:
    user_input = input("Enter a nested list (e.g., [[1,2],[2,3],[2,4]]): ")

    import ast
    nested_list = ast.literal_eval(user_input)

    elem_input = input("Enter the element to count: ")
    try:
        elem = int(elem_input)
    except ValueError:
        try:
            elem = float(elem_input)
        except ValueError:
            elem = elem_input

    result = count_occurrences(nested_list, elem)

    print("Nested List:", nested_list)
    print(f"Occurrences of {elem}:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list.")
```

##7. Flatten a List of Lists of Lists:


```python
def deep_flatten(lst):
    """
    Recursively flattens a deeply nested list into a single list.

    Parameters:
    lst (list): The nested list of any depth.

    Returns:
    list: A single flattened list containing all elements.
    """
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)
    return flat_list
try:
    user_input = input("Enter a deeply nested list (e.g., [[[1,2],[3,4]],[[5,6],[7,8]]]): ")

    import ast
    nested_list = ast.literal_eval(user_input)

    result = deep_flatten(nested_list)

    print("Nested List:", nested_list)
    print("Flattened List:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list.")
```

##8. Nested List Average:


```python
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): The nested list containing numbers and/or sublists.

    Returns:
    float: The average of all elements in the nested list.
    """
    def flatten(l):
        """Helper function to flatten nested list of any depth."""
        flat = []
        for item in l:
            if isinstance(item, list):
                flat.extend(flatten(item))
            else:
                flat.append(item)
        return flat

    flat_list = flatten(lst)
    if len(flat_list) == 0:
        return 0
    return sum(flat_list) / len(flat_list)
try:
    user_input = input("Enter a nested list (e.g., [[1,2],[3,4],[5,6]]): ")

    import ast
    nested_list = ast.literal_eval(user_input)

    result = average_nested(nested_list)

    print("Nested List:", nested_list)
    print("Average of all elements:", result)

except (SyntaxError, ValueError):
    print("Error: Please enter a valid nested list containing numbers.")
```
